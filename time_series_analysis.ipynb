# Decomposition Using a Hodrick-Prescott Filter

 Using a Hodrick-Prescott Filter, decompose the exchange rate price into trend and noise.

```python
# Import statsmodels
import statsmodels.api as sm

# Create the new DataFrame by stripping out O, H and L for adding to, later
cadjpy_trnd_nse = cad_jpy_df.drop(columns={'Open','High','Low'})

# Apply the Hodrick-Prescott Filter by decomposing the exchange rate price into two separate series:
cad_jpy_noise, cad_jpy_trend = sm.tsa.filters.hpfilter(cadjpy_trnd_nse)
```

```python
# Create a dataframe of just the exchange rate price, and add columns for "noise" and "trend" series from above:
# Taking the DataFrame created above ...
cadjpy_trnd_nse['noise'] = cad_jpy_noise
cadjpy_trnd_nse['trend'] = cad_jpy_trend
cadjpy_trnd_nse.head()
```

```python
# Plot the Exchange Rate Price vs. the Trend for 2015 to the present
# No need to create a new DataFrame with the column dropped
# Filter time-series data easily with loc.
cadjpy_trnd_nse.loc["2015":].drop(columns='noise').plot(title="Price vs. Trend", figsize=(14,8))
```
